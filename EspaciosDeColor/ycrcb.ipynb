```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
ballenita = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI2\rubberwhale1.png")
plt.imshow(ballenita)
```

```python
ballena1Y = cv.cvtColor(ballenita, cv.COLOR_BGR2YCrCb)
y, cr, cb = cv.split(ballena1Y)

canal_y_eq = cv.equalizeHist(y)

img_ycrcb = cv.merge([canal_y_eq, cr, cb])

ballenita_final = cv.cvtColor(img_ycrcb, cv.COLOR_YCrCb2RGB)

plt.imshow(ballenita_final)
```

```python
#Apilar
# Imagen original e imagen unida con el canal ecualizado
plt.imshow(np.hstack((ballenita[:,:,::-1], ballenita_final)))
```

## HSL

```python
#Ballena en HLS
ballena1 = cv.cvtColor(ballenita, cv.COLOR_BGR2HLS)

#Separar canales de HLS
y, cr, cb = cv.split(ballena1)

#Ecualizar canal y
canal_y_eq = cv.equalizeHist(y)

#Mergear canal y ecualizado y los otros 2 intactos
img_ycrcb = cv.merge([canal_y_eq, cr, cb])

#Convertir de HLS a RGB
ballenita_final = cv.cvtColor(img_ycrcb, cv.COLOR_HLS2RGB)

#Plotear
plt.imshow(np.hstack((ballenita[:,:,::-1], ballenita_final)))
```

```python
#Ballena en HLS
ballena1 = cv.cvtColor(ballenita, cv.COLOR_BGR2HLS)

#Separar canales de HLS
y, cr, cb = cv.split(ballena1)

#Ecualizar canal y
canal_cr_eq = cv.equalizeHist(cr)

#Mergear canal y ecualizado y los otros 2 intactos
img_ycrcb = cv.merge([y, canal_cr_eq, cb])

#Convertir de HLS a RGB
ballenita_final = cv.cvtColor(img_ycrcb, cv.COLOR_HLS2RGB)

#Plotear
plt.imshow(np.hstack((ballenita[:,:,::-1], ballenita_final)))
```
