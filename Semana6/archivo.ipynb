```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Imagen blanca
#PARAMETROS: Tamaño de la matriz, tipo de dato (hay que cambiarlo a int ya que está predeterminado como float)
chl1 = np.ones((300,300), dtype=np.uint8)*255
```

```python
#Imagenes negras
chl2 = np.zeros((300,300), dtype=np.uint8)
chl3 = np.zeros((300,300), dtype=np.uint8)
```

```python
#Merge = unir
#NOTA: Las tres matrices deben de tener la misma dimension
imgRoja = cv.merge((chl1, chl2, chl3))
```

```python
plt.imshow(imgRoja)
```

```python
#[:, :, :: -1] Convertir de RGB a BGR
cv.imwrite("roja.jpg", imgRoja [:, :, :: -1])
```

```python
chl1a = np.ones((300,300), dtype=np.uint8)*255
chl2a = np.ones((300,300), dtype=np.uint8)*255
chl3a = np.zeros((300,300), dtype=np.uint8)
```

```python
imgAmarilla = cv.merge((chl1a, chl2a, chl3a))
```

```python
plt.imshow(imgAmarilla)
```

```python
cv.imwrite("Amarillo.jpg", imgAmarilla [:, :, :: -1])
```
