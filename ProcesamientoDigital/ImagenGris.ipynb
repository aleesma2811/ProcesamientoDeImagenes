```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
windows_logo = cv.imread(r"c:/Users\Ale Escallada/OneDrive\Documentos/imagenesPI\WindowsLogo.jpg",0)
windows_logo
```

```python
filas, columnas = windows_logo.shape
```

```python
img_resultante = windows_logo.copy() #Copia del objeto para no modificar el original
```

```python
for f in range(filas):
    for c in range(columnas):
        if (windows_logo[f,c] < 128):
            img_resultante[f,c] = 0
        if (windows_logo[f,c] >= 128):
            img_resultante[f,c] = 255
```

```python
plt.imshow(img_resultante, cmap="gray")
```
