```python
#[:, :, :: -1] Cambiar de 
img = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI\apple.jpg")[:, :, :: -1]
img.shape
```

```python
plt.imshow(img)
```

```python
#Separar canales en RGB
canal_R = img[:, :, 0]
canal_G = img[:, :, 1]
canal_B = img[:, :, 2]
```

```python
canal_B.shape
```

```python
#Split divide espacios en colores
canal_R, canal_G, canal_B = cv.split(img)
```

```python
plt.imshow(canal_B)
```

```python
#Canal azul representado en escala de grises
plt.imshow(canal_B, cmap="gray")
```

```python
#Canal rojo
plt.imshow(canal_R)
```

```python
#Canal rojo en escalas de grises
plt.imshow(canal_R, cmap="gray")
plt.title("Canal rojo")
```

```python
#Se ejecutan una tras otra, entonces plt.show() muestra cada una
plt.imshow(canal_R, cmap="gray")
plt.title("Canal rojo")
#plt.show()
plt.imshow(canal_G, cmap="gray")
plt.title("Canal verde")
#plt.show()
plt.imshow(canal_B, cmap="gray")
plt.title("Canal azul")
#plt.show()
```

```python
#(1 fila, 3 columnas, elemento 1)

plt.subplot(131); plt.imshow(canal_R, cmap="gray"); plt.title("Canal rojo")
plt.subplot(132); plt.imshow(canal_G, cmap="gray"); plt.title("Canal verde")
plt.subplot(133); plt.imshow(canal_B, cmap="gray"); plt.title("Canal azul")
#La imagen tiene muy poco azul ya que hay mucho negro
#Canal rojo, el rojo predomina ya que la mayora de los pixeles son blancos
```
