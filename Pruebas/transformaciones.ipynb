```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
ardilla = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI\squirrel_cls.jpg")[:, :, :: -1]
plt.imshow(ardilla)
```

```python
ardilla_gris = cv.cvtColor(ardilla, cv.COLOR_BGR2GRAY)
plt.imshow(ardilla_gris, "gray")
```

### Logarítmica

```python
m = np.max(ardilla_gris)
m
```

```python
c = 255/(np.log(1 + m))
c
```

```python
ardilla_res = ardilla_gris.copy()
filas, columnas = ardilla_gris.shape

for f in range(filas):
    for col in range(columnas):
        ardilla_res[f, col] = c*np.log(1 + ardilla_gris[f, col])

plt.imshow(ardilla_res, "gray")
```

```python
plt.imsave("ardillaBin.jpg", ardilla_res, cmap="gray")
```
