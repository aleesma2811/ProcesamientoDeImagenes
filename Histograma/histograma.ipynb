### Ecualizar/Histograma de imagen a color

```python
figuras = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI2\pic3.png")[:, :, ::-1]
plt.imshow(figuras)
```

```python
figuras.size
```

```python
b, g, r = cv.split(figuras)
```

```python
img_eq_b = cv.equalizeHist(b)
img_eq_g = cv.equalizeHist(g)
img_eq_r = cv.equalizeHist(r)

img_final_eq = cv.merge([img_eq_r, img_eq_g, img_eq_b]) #Cambiar de BGR a RGB

img_final = cv.calcHist([img_final_eq], [0], None, [256], (0, 256))

plt.plot(img_final)
```

```python
plt.imshow(img_final_eq)
```

```python
#Histograma de cada color
plt.plot(cv.calcHist([img_eq_b], [0], None, [256], (0, 256)), "b")
plt.plot(cv.calcHist([img_eq_g], [0], None, [256], (0, 256)), "g")
plt.plot(cv.calcHist([img_eq_r], [0], None, [256], (0, 256)), "r")
```

### Ecualizar sin RGB

Leer imagen a color
COnvertir espacio de color (YCrCb)
Separar los canales
Ecualizar el canal Y
Merge a los canales
Convertir de YCrCb a RGB
Plot

```python
ballenita = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI2\rubberwhale1.png")
plt.imshow(ballenita)
```

```python
cv.cvtColor(ballenita, cv.COLOR_BGR2YCrCb)
y, cr, cb = cv.split(ballenita)

canal_y_eq = cv.equalizeHist(y)

img_ycrcb = cv.merge([canal_y_eq, cr, cb])

ballenita_final = cv.cvtColor(img_ycrcb, cv.COLOR_YCrCb2RGB)

plt.imshow(ballenita_final) 
```
