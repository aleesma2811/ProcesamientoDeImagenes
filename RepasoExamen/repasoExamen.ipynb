```python
ardilla = cv.imread(r"c:\Users\Ale Escallada\OneDrive\Documentos\imagenesPI\squirrel_cls.jpg")[:, :, :: -1]
```

```python
plt.imshow(ardilla)
```

```python
#Separar canales en RGB
canal_R = ardilla[:, :, 0]
canal_G = ardilla[:, :, 1]
canal_B = ardilla[:, :, 2]
```

```python
canal_R, canal_G, canal_B = cv.split(ardilla)
```

```python
plt.subplot(1,3,1); plt.imshow(canal_R); plt.title("Canal rojo")
plt.subplot(1,3,2); plt.imshow(canal_G); plt.title("Canal Verde")
plt.subplot(1,3,3); plt.imshow(canal_B); plt.title("Canal Azul")
```

```python
plt.subplot(1,3,1); plt.imshow(canal_R, cmap="gray"); plt.title("Canal rojo")
plt.subplot(1,3,2); plt.imshow(canal_G, cmap="gray"); plt.title("Canal Verde")
plt.subplot(1,3,3); plt.imshow(canal_B, cmap="gray"); plt.title("Canal Azul")
```
